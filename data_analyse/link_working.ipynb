```python
def text_month_to_number(text):
    if text == 'January':
        return '1'
    elif text == 'February':
        return '2'
    elif text == 'March':
        return '3'
    elif text == 'April':
        return '4'
    elif text == 'May':
        return '5'
    elif text == 'June':
        return '6'
    elif text == 'July':
        return '7'
    elif text == 'August':
        return '8'
    elif text == 'September':
        return '9'
    elif text == 'October':
        return '10'
    elif text == 'November':
        return '11'
    elif text == 'December':
        return '12'
    else:
        raise Exception("invalid month format")
    
```

```python
df = pd.read_csv('pages.csv')
for i in range(df["CTR"].count()):
    df.at[i, 'CTR'] = float(df.at[i, 'CTR'].replace('%', ''))
df.fillna(0, inplace=True)
df = df.sort_values(by='Impressions')
```

```python
soups = [None for x in range(len(df['Top pages']))]
i = 0
for url in tqdm(df["Top pages"]):
    reqs = requests.get(url)
    soups[i] = BeautifulSoup(reqs.text, 'html.parser')
    i += 1
```

```python
titr = []
lead = []
date = []
week_day = []
time = []
tags_all = []
image_count = []

i = 0
for soup in soups:
    url = df['Top pages'][i]
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')


    date_list = list(filter(None, soup.find('span', {'class':"en_date visible-lg visible-md"}).string.replace('\t', '').split(' ')))
    date_list[1] = text_month_to_number(date_list[1])
    date_list = [int(x) for x in date_list]
    date.append(date_list)
    week_day.append((calendar.weekday(date_list[2], date_list[1], date_list[0]) + 2) % 7 + 1)


    titr.append(soup.find('h1', {'class':"Htag"}).string)

    lead.append(soup.find('div', {'class':'subtitle'}).get_text() if soup.find('div', {'class':'subtitle'}) != None else "")
    

    persian_time = (soup.find('span', {'class':'fa_date'}).get_text().split(' ')[4].split(':'))
    persian_time = [int(unidecode(x)) for x in persian_time]
    # print(persian_time)
    time.append(persian_time)
    
    tags = [tag.replace(' ', '').replace('\t', '').replace('\xa0', '') for tag in soup.find('div', {'class':'tag_items'}).strings]
    tags = list(filter(None, tags))
    tags_all.append(tags)


    # images = soup.body.find('div', {'class':re.compile(r".*col-md-22.*$")}).find_all('img', {'class':re.compile(r"^.*img-responsive.*$")})
    # print(images)
    print(url)
    # for image in soup.body.find('div', {'class':re.compile(r".*col-md-22.*$")}).find_all('img'):
    #     print(image)

    
    i += 1

print(titr)
print(lead)
print(date)
print(time)
print(tags_all)
```
